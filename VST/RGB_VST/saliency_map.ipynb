```python
import shutil
import glob
import os

# Specify the source and destination directories
source_directory = '/home/hp/Downloads/images'  # Replace with your source directory path
destination_directory = '/home/hp/sam/VST/RGB_VST/Data/DUTS/DUTS-TE/DUTS-TE-Image'  # Replace with your destination directory path

# Define the file extensions for image files
image_extensions = ['*.jpg', '*.jpeg', '*.png', '*.gif', '*.bmp', '*.tiff']

def copy_images(source, destination):
    """Copy all images from source to destination."""
    image_files = []
    for ext in image_extensions:
        image_files.extend(glob.glob(os.path.join(source, ext)))

    for image_file in image_files:
        try:
            shutil.copy(image_file, destination)
            print(f'Copied {image_file} to {destination}')
        except Exception as e:
            print(f'Error copying {image_file}: {e}')

def delete_images(directory):
    """Delete all images in the directory."""
    image_files = []
    for ext in image_extensions:
        image_files.extend(glob.glob(os.path.join(directory, ext)))

    for image_file in image_files:
        try:
            os.remove(image_file)
            print(f'Deleted {image_file}')
        except Exception as e:
            print(f'Error deleting {image_file}: {e}')

# Get subdirectories and sort them
subdirectories = sorted(next(os.walk(source_directory))[1])

# Iterate over each subfolder in the sorted order
for subdir in subdirectories:
    source_subdir = os.path.join(source_directory, subdir)
    print(f'Processing folder: {source_subdir}')
    copy_images(source_subdir, destination_directory)

    # Run your processing script
    !python train_test_eval.py --Testing True

    # Delete the copied images from the destination directory
    delete_images(destination_directory)
```
